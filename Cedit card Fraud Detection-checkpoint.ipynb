```python
#determining the number of fraud cases

fraud = df[df['Target']==1]
valid = df[df['Target']==0]
outlierFraction = len(fraud)/float(len(valid))
print(outlierFraction)
```

```python
print('Fraud Cases: {}' .format(len(df[df['Target']==1])))
```

```python
print('Valid Transactions: {}' .format(len(df[df['Target']==0])))
```
